Figure 2 | The FTO obesity risk haplotype derepresses IRX3 and IRX5 expression in adipocytes. 

a,
**Upper panel**: Regional LD plot showing the 2.5Mb genomic interval spanning an  extended region of the obesity association region of high LD in intron 1 and 2 of FTO. The variants surrounding the GWAS-reported tagSNP rs1558902 are color coded to reflect their LD with rs1558902 <strike>**in Europeans**</strike> (pairwise $r^2$ values from 1000 Genomes).

**Lower panel**: Higher-order chromatin interactions revealed by Hi-C assays and linkage disequilibrium r2 statistics at 40kb-resolution. The upper triangle pixels represent Hi-C contact  probabilities from human IMR90 myofibroblastic cells **{Rao, 2014}** computed on 5kb genomic bins (matrix balancing done as in {Imakaev,2012}). They reveal a discrete Topologically Associating Domain (TAD) in the obesity association region containing FTO, RFGRIP1L, IRX3, IRX5, CRNDE and IRX6. The lower triangle contains pairwise $r^2$ data on genetic variants <strike>**in Europeans**</strike>. All-by-all pairwise Pearson $r^2$ values were calculated on phase 1 variant data from the 1000 Genomes Project. Pixel color corresponding to a pair of genomic bins reflects the **base-10 logarithm of the mean $r^2$ value for all pairs of variants, one from each bin**.

    b, Haplotype-specific relative gene expression of target genes in primary human adipose stromal cells (hASC) from individuals carrying the homozygous non-risk (n = 24) or the risk haplotype  (n = 24) were cultured, induced to differentiate into adipocytes, and harvested after 2 days. mRNA levels of target genes were measured by qPCR and normalised to HPRT mRNA. P-values were calculated by t-test, error bars show SD.


```python
from __future__ import division
%matplotlib inline
import matplotlib as mpl
import seaborn as sns
sns.set_style('white') 
sns.set_context('paper') #poster, talk

import matplotlib.pyplot as plt
import numpy as np
import datetime
import os
pjoin = os.path.join
```

```python
PROJECT_DIR = pjoin(os.getcwd(), os.path.pardir)
BINSIZE = 5000
START = 52800000//BINSIZE
STOP  = 56000000//BINSIZE
CHROMLEN = 90354753
```

```python
# utils...
def reveal_triu(A):
    B = np.array(A)
    B[np.tril_indices_from(B)] = np.nan
    return B

def reveal_tril(A):
    B = np.array(A)
    B[np.triu_indices_from(B)] = np.nan
    return B

def chx(start, **kw):
    kw.setdefault('light', 1.0)
    kw.setdefault('dark', 0.0)
    kw.setdefault('as_cmap', True)
    return sns.cubehelix_palette(start=start, **kw)

class Binner(object):
    def __init__(self, bp_start, bp_end, bp_step):
        import math
        self.start, self.stop, self.step = bp_start, bp_end, bp_step
        self.n_bins = int(math.floor((bp_end - bp_start)/bp_step))
    
    def get(self, bp_coord):
        i = (bp_coord - self.start)//self.step
        return max(0, min(i, self.n_bins-1))
binner = Binner(0, CHROMLEN, BINSIZE)
```

```python
# Get Hi-C data
#/archive/compbio/users/nezar/current/melina/data/04-hic/IMR90.MAPQGE30.chr16_5kb.ICEobserved.npy
chr16 = np.load(pjoin(PROJECT_DIR, 'data/04-hic/IMR90.MAPQG0.chr16_5kb.ICEobserved.npy'))
colsum = chr16.sum(axis=0).max()
A = chr16[START:STOP, START:STOP]
N = len(A)
```

```python
# Get LD data
L = np.loadtxt(pjoin(PROJECT_DIR, 'data/03-ld-aggregate/chr16.52800000-56000000.ld.5000.mean.txt.gz'))
```

```python
names = [
    'rs1421085',
    'IRX5',
    'IRX3',
    'FTO',
    'RPGRIP1L',
    'RBL2',
    'CHD9',
    'IRX6',
    'CRNDE',
]

locations = [ #bp coordinate
    (53800954, 53800954), # SNP rs1421085
    (54964200, 54965302),
    (54319016, 54319016),
    (53737375, 53738194),
    (53737375, 53738194),
    (53467832, 53468474),
    (53088445, 53089051),
    (55357172, 55357772),
    (54973696, 54974296),
]

name_point = { #bin coordinate
    name: (binner.get(loc[0]), binner.get(loc[1])) \
        for (name, loc) in zip(names, locations)
}
```

```python
from matplotlib.ticker import MaxNLocator, FuncFormatter

# 1 main axis + 2 colorbar axes
f, (ax, cax1, cax2) = plt.subplots(3,1, figsize=(20,50))

# plot HiC/LD
imA = ax.matshow(reveal_triu(np.log(A)), cmap=chx(0.6), interpolation='none')
imL = ax.matshow(reveal_tril(np.log10(L)), cmap=plt.cm.Blues, interpolation='none')
ax.set_xlim([0, N])
ax.set_ylim([N, 0])
bbox = ax.get_position()

# genomic coordinate ticks
tick_locator = MaxNLocator(4)
tick_formatter = FuncFormatter( lambda x, pos: '{:7.0f}'.format(START*BINSIZE + x*BINSIZE) )
ax.xaxis.set_major_locator(tick_locator)
ax.xaxis.set_major_formatter(tick_formatter)
ax.yaxis.set_major_locator(tick_locator)
ax.yaxis.set_major_formatter(tick_formatter)

#Hi-C colorbar (units = contact probability)
cax1.set_position([
    bbox.x0+bbox.width-0.07, 
    bbox.y0, 
    0.015, 
    bbox.height,
])
cbar = f.colorbar(imA, cax=cax1)
cmin, cmax = cbar.get_clim()
cticks = np.linspace(cmin, cmax, 5)
cbar.set_ticks(cticks)
cbar.set_ticklabels(['{:0.2f}'.format(np.exp(t)/colsum * 100) for t in cticks])

# LD colorbar (units = log10 r^2-value)
cax2.set_position([
    bbox.x0+bbox.width,
    bbox.y0,
    0.015,
    bbox.height,
])
cbar = f.colorbar(imL, cax=cax2)
cmin, cmax = cbar.get_clim()
cticks = np.linspace(cmin, cmax, 5)
cbar.set_ticks(cticks)
cbar.set_ticklabels(['{:0.1f}'.format(t) for t in cticks])

# identify loci
for name, point in name_point.items():
    a, b = point
    kw = {}
    if name == 'rs1421085':
        kw['marker'] = 'o'
        kw['markerfacecolor'] = 'r'
        kw['markeredgecolor'] = 'r'
        kw['markeredgewidth'] = 1
    else:
        kw['marker'] = 'o'
        kw['markerfacecolor'] = 'none'
        kw['markeredgecolor'] = 'r'
        kw['markeredgewidth'] = 1
    ax.plot(a-START, a-START, **kw)

    
p = name_point['CHD9'][0]-START; d = 160
ax.plot([p, p], [p, p+d], 'k--', linewidth=1)
ax.text(p, p+d+2, 'CHD9')

p = name_point['RBL2'][0]-START; d = 160
ax.plot([p, p], [p, p+d], 'k--', linewidth=1)
ax.text(p, p+d+2, 'RBL2')

p = name_point['FTO'][0]-START; d = 320
ax.plot([p, p], [p, p+d], 'k--', linewidth=1)
ax.text(p, p+d+4, 'FTO\nRFGRIP1L')

p = name_point['rs1421085'][0]-START; d = 160
ax.plot([p, p], [p, p+d], 'k--', linewidth=1)
ax.text(p, p+d+2, 'rs1421085')

p = name_point['IRX3'][0]-START; d = 160
ax.plot([p, p], [p, p+d], 'k--', linewidth=1)
ax.text(p, p+d+2, 'IRX3')

p = name_point['IRX5'][0]-START; d = 160
ax.plot([p, p], [p, p+d], 'k--', linewidth=1)
ax.text(p, p+d+4, 'IRX5\nCRNDE')

p = name_point['IRX6'][0]-START; d = 80
ax.plot([p, p], [p, p+d], 'k--', linewidth=1)
ax.text(p, p+d+2, 'IRX6');

#plt.savefig(pjoin(PROJECT_DIR, 'publish/new_{}.png'.format(datetime.date.today())), bbox_inches='tight', pad_inches=0)
```
