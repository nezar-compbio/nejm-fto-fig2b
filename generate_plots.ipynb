```python
from hictools.settings import CHROM_LENS
from hictools.binning import Binner

binner = Binner(0, CHROM_LENS[15], binsize=40000)

names = [
    'rs1421085',
    'IRX5',
    'IRX3',
    'FTO',
    'RPGRIP1L',
    'RBL2',
    'CHD9',
    'IRX6',
    'CRNDE',
]

locations = [
    '53800954',
    '54964200-54965302',
    '54319016-54319016',
    '53737375-53738194',
    '53737375-53738194',
    '53467832-53468474',
    '53088445-53089051',
    '55357172-55357772',
    '54973696-54974296',
]

colors = [
    '#000000',
    '#E41A1C',
    '#377EB8',
    '#4DAF4A',
    '#984EA3',
    '#FF7F00',
    '#FFFF33',
    '#A65628',
    '#F781BF',
]
```

```python
from matplotlib.ticker import MaxNLocator, FuncFormatter

norm = 1960
vmax = (chr16/norm).max()
plt.matshow(
    np.log(chr16/norm),
    cmap=plt.cm.PuBu,
    vmax=np.log(vmax))
lims = [1320, 1399]
plt.xlim(lims)
plt.ylim(lims[::-1])


# relabel axes
ax = plt.gca()
tick_locator = MaxNLocator(4)
tick_formatter = FuncFormatter( lambda x, pos: '{:7.0f}'.format(x*40000) )
ax.xaxis.set_major_locator(tick_locator)
ax.xaxis.set_major_formatter(tick_formatter)
ax.yaxis.set_major_locator(tick_locator)
ax.yaxis.set_major_formatter(tick_formatter)

#colors = dark2_colors
#colors[0] = '#000000'
# loci
seen = set()
for i in reversed(range(len(names))):
    sel = binner.select(locations[i])
    kw = {'marker': 's', 
          'markeredgecolor':'none'}
    if i==0:
        kw['marker'] = 'o'
        kw['markerfacecolor'] = colors[i]
        #kw['markeredgecolor'] = colors[i]
    elif sel[0] in seen:
        kw['marker'] = 's'
        kw['markerfacecolor'] = 'none' #colors[i]
        kw['markeredgecolor'] = colors[i]
        kw['markeredgewidth'] = 3
    plt.plot(sel, sel, **kw)
    seen.add(sel[0])

# colorbar
cbar = plt.colorbar(fraction=0.04539)
cmin, cmax = cbar.get_clim()
cticks = np.linspace(cmin, cmax, 5)
cbar.set_ticks(cticks)
cbar.set_ticklabels(['{:0.3f}'.format(np.exp(t)) for t in cticks])

# legend
leg = plt.legend(ax.lines[::-1], names, bbox_to_anchor=(1.5, 1.0), numpoints=1)
```
